## Stacking

```python
## In Horizontal Stacking rows need to be equal to each other
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(15).reshape(3,5)
np.hstack((a4,a5))    
```

```python
## In Vstack columns needs to be equal
a6 = np.arange(12).reshape(4,3)
a7 = np.arange(15).reshape(5,3)
np.vstack((a6,a7))
```
