```python
# pehle dekh le kaunse matches the
no_result_matches = ipl[ipl['WinningTeam'].isna()]
no_result_matches
```

```python
# ab clean dataframe banale (sirf wahi matches jinka result tha)
ipl_clean = ipl.dropna(subset=['WinningTeam'])
```

```python
ipl_clean
```

```python
match_played = ipl_clean['Team1'].value_counts() + ipl_clean['Team2'].value_counts()
wins = ipl_clean['WinningTeam'].value_counts()
losses = match_played.subtract(wins,fill_value = 0)
```

```python
summary = pd.DataFrame({
    'Total Matches' : match_played,
    'Wins' : wins,
    'Losses' : losses
})
summary.sort_values(by = 'Total Matches')
```
