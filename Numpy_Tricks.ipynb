## np.sort

```python
## It returns an sorted copy of an array
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
np.sort(a)
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(b) ## by default it sort our array row wise
```

```python
## If we want to sort our array column wise so give write axis = 0
np.sort(b, axis = 0)
```

```python
## if we want to sort descending wise so we need to write [::-1]
np.sort(b, axis = 0)[::-1]
```

```python
## we dont use sorted function because it gives us python list and np.sort gives us array, thats why we will use np.sort
```

# np.concatenate()
## This function concatenate a sequence of arrays along an existing axis 

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
np.concatenate((c,d),axis = 0)
```

```python
# Rule 1: Arrays ki dimensions same honi chahiye
# Rule 2: axis=0 par columns same honi chahiye
# Rule 3: axis=1 par rows same honi chahiye
```
