# Pandas Series?
## "Pandas Series is a one-dimensional labeled array capable of storing any data type and having an associated index."

## Series = Single Column Data
## DataFrame = Multiple Series ka collection.

```python
import pandas as pd
import numpy as np
```

## Create Series using python list.

```python
country = ["India", "Pakistan", "Srilanka", "USA"]
pd.Series(country)
```

```python
country[2]
```

```python
batsman_run = [86,76,23,59,90]
pd.Series(batsman_run, dtype = 'int32')
```

## Custom Index

```python
marks = [56,78,45,90]
subjects = ['Science','English','Hindi','Maths']
pd.Series(marks, index =subjects)
```

```python
## Setting a Name
ad_marks = pd.Series(marks, index =subjects,name = "Adarsh's Marks")
```

```python
ad_marks
```

## Create Series using dictionary

```python
marks = {
    "Maths":90,
    "Hindi":78,
    "English": 87,
    "Science":67
}

marks_series = pd.Series(marks, name = "Adarsh's Marks")
marks_series
```

## Series Attributes

```python
marks_series.index ## it shows us the indexes
```

```python
marks_series.is_unique ## It tells us that our values are unique or not
```

```python
marks_series.values ## it shows us the values in the form of numpy array
```

```python
marks_series.ndim
```
